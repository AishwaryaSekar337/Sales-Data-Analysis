```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Online Sales Data.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.replace(r'^\s+',np.nan, regex= True, inplace=True)
df.isnull()
df.head(20)
```

```python
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include='number').columns
df
```

```python
numberic_cols = ["Units Sold", "Unit Price", "Total Revenue"]
```

```python
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
    print(df.head(5))
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
print(df.head(10))
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
print(df)
```

```python
for col in df.columns:
    print(repr(col))
```

```python
df["Total Revenue"] = (df["Units Sold"]
* df["Unit Price"])
print(df)
```

```python
top_revenue_products = df.groupby("Product Name")["Total Revenue"].sum().sort_values(ascending=False).head(10)
print(top_revenue_products)
top_products = df.groupby("Product Name")["Units Sold"].sum()
print(top_products.sort_values(ascending=False))
```

```python
print(df["Date"].unique())
```

```python
df["Date"] = pd.to_datetime(df["Date"],
                            dayfirst=True)
print(df)
```

```python
df["Month"] = df["Date"].dt.month
Monthly_sales = df.groupby("Month")
Monthly_sales = Monthly_sales["Total Revenue"].sum()
print(Monthly_sales)
```

```python
category_sales =df. groupby("Product Category")["Total Revenue"].sum()
print(category_sales)
```

```python
Region_sales = df.groupby("Region")["Total Revenue"].sum()
print(Region_sales)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
monthly_sales = df.groupby("Month")["Total Revenue"].sum()
plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.grid()
plt.show()
```

```python
top_products = df.groupby("Product Name")["Units Sold"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
plt.bar(top_products.index, top_products.values)
plt.title("Top 10 Products by Units Sold")
plt.xlabel("Product Name")
plt.ylabel("Units Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = df.groupby("Region")["Total Revenue"].sum()
plt.figure(figsize=(6,6))
plt.pie(region_sales.values, labels=region_sales.index, autopct='%1.1f%%')
plt.title("Revenue by Region")
plt.show()
```

```python
payment_counts = df["Payment Method"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(payment_counts.index, payment_counts.values)
plt.title("Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
category_sales = df.groupby("Product Category")["Total Revenue"].sum()
category_sales.plot(kind="bar")
plt.title("Revenue by Product Category")
plt.show()
```

```python
print("Total Revenue =", df["Total Revenue"].sum())

print("Best Product =",
      df.groupby("Product Name")["Total Revenue"].sum().idxmax())

print("Best Region =",
      df.groupby("Region")["Total Revenue"].sum().idxmax())

print("Best Payment Method =",
      df.groupby("Payment Method")["Total Revenue"].sum().idxmax())
```

```python
upi_sales = df[df["Payment Method"] == "UPI"]

print(upi_sales)
```

```python
high_revenue = df[df["Total Revenue"] > 1000]

print(high_revenue)
```
